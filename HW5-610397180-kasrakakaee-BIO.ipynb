```python
from mpmath import mp
import mpmath
from decimal import Decimal
import numpy as np
import matplotlib.pyplot as plt
import random
mp.dps = 50
```

```python
#-10<=x,y<=10
def f(x, y):
  return abs(np.sin(x) * np.cos(y) * np.exp(abs(1 - np.sqrt(x**2 + y**2) / np.pi)))
```

```python
#-100<=x,y<=100
def g(x, y):
  return x * (np.sin(np.pi * np.cos(x) * np.tan(y)) * np.sin(y/x) / (1 + np.cos(y/x)))
```

```python
def initialize(n, r, ds):
  X = np.random.rand(2, n) * 2 * r - r
  X = np.longdouble(X)
  V = np.random.randn(2, n) * ds
  V = np.longdouble(V)
  return X, V
```

```python
def solve1(n, r, ds, c1, c2, w, iter_count):
  X, V = initialize(n, r, ds)
  local_best = X
  local_best_val = g(X[0], X[1])
  global_best = local_best[:, local_best_val.argmin()]
  global_best_val = local_best_val.min()

  for _ in range(iter_count):
    r = np.random.rand(2)
    V = w * V + r[0] * c1 * (local_best - X) + r[1] * c2 * (global_best.reshape(-1, 1) - X)
    X = X + V
    new_val = g(X[0], X[1])
    local_best[:, (local_best_val >= new_val)] = X[:, (local_best_val >= new_val)]
    local_best_val = np.array([local_best_val, new_val]).min(axis = 0)
    global_best = local_best[:, local_best_val.argmin()]
    global_best_val = local_best_val.min()
    print(global_best_val)
  print(global_best)
  print(g(global_best[0], global_best[1]))
```

```python
solve1(500, 100, 1, 0.1, 0.1, 0.8, 500)
```

```python
def solve2(n, r, ds, c1, c2, w, iter_count):
  X, V = initialize(n, r, ds)
  local_best = X
  local_best_val = f(X[0], X[1])
  global_best = local_best[:, local_best_val.argmax()]
  global_best_val = local_best_val.max()

  for _ in range(iter_count):
    r = np.random.rand(2)
    V = w * V + r[0] * c1 * (local_best - X) + r[1] * c2 * (global_best.reshape(-1, 1) - X)
    X = X + V
    new_val = f(X[0], X[1])
    local_best[:, (local_best_val <= new_val)] = X[:, (local_best_val <= new_val)]
    local_best_val = np.array([local_best_val, new_val]).max(axis = 0)
    global_best = local_best[:, local_best_val.argmax()]
    global_best_val = local_best_val.max()
    print(global_best_val)
  print(global_best)
  print(f(global_best[0], global_best[1]))
```
