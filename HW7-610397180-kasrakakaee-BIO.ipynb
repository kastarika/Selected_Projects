```python
import numpy as np
```

tabee dist ke faseleye do noghte ra baraye djo arayeye a va b barmigardanad

```python
def dist(a, b):
  return np.linalg.norm(a - b, axis=1)
  #return np.sqrt(np.square(a[:,0] - b[:,0]) + np.square(a[:,1] - b[:,1]))
```

init neural network ra misazad. yani ba tedade delkhah vazn ha ra random meghdar dehi mikonad dar yek network 2 * size

```python
def init(city_count):
  return np.random.rand(city_count, 2)
```

get cities file e voroodi ra mikhanad va etelaat ra dar cities mirizad va barmigardanad

```python
def get_cities(file_path):
  f = open(file_path, "r")
  cities = []
  lines = f.readlines()
  for line in lines:
    inp = list(map(np.float64,line.split()))
    cities.append(inp)
  f.close()

  return np.array(cities)
```

check neighbor faseleye d va radius ra migirad va be ezaye node i ke kamtar az radius fasele darad 1 va agar bishtar fasele darad 0 barmigardanad. dar nahayat be komake in tabe va numpy araye ye 0 o 1 i ijad mikonim ke moshakhas mikonad che node hayee dar hamsayegie node i ke taghir mikonad ast va bayad taghir konad 


```python
def check_neighbor(d, radius):
  if(d <= radius):
    return 1
  else:
    return 0
```

find neighborhood node i ke entekhab shode ra migirad va araye i ke moshakhas mikonad che node hayee dar hamsayegi radius an gharar darad ra be soorate 0 o 1 barmigardanad. 

```python
def find_neighborhood(node, radius, size):
  radius = max(radius, 1)
  d = np.absolute(np.arange(size) - node)
  d = np.minimum(d, size - d)
  cn = np.vectorize(check_neighbor)
  return cn(d, radius)
```

find path ba tavajoh be neural networki ke darim haman tori ke tozih dadim masiri ke in network peida karde ast ra barmigardanad.(be tartib poshte sare ham)

```python
def find_path(nn, cities, real_cities):
  order = []
  for city in cities:
    c = city[1:3]
    best = dist(nn, c).argmin()
    order.append([best,city[0] - 1])

  order.sort()
  path = []
  for city in order:
    path.append([ real_cities[int(city[1])][1], real_cities[int(city[1])][2]])

  return path
```

eval path adade meghadre masiri ke path tey mikonad ra barmigardanad

```python
def eval_path(path):
  return np.sum(dist(path, np.roll(path, 1, axis = 0)))
```

solve avval neural network nn ra be size e 10 barabare tedad shahr ha ba vazn haye random misazad. sepas mokhtasate shahr ha ra normalize mikonad va copy asli ra negah midadarad. sepas dar har iteration avval shahr ha ra shuffle mikonad. sepas baraye set e shahr ha kohonen ra anjam midahad. dar har bar learn kardane yek shahr learning rate va hamsayegi ra koochik tar mikonim.
zamani ke learning rate ya hamsayegi kheili koochik tar shod yani digar learning e ghabele tavajohi etefagh nemioftad pas alogrithm ra motevaghef mikonim

```python
def solve(cities, iteration_count, learning_rate):
  size = len(cities)
  n = size * 10
  nn = init(n)
  
  real_cities = cities.copy()
  mn = np.amin(cities, axis = 0)
  mn = min(mn[1], mn[2])
  mx = np.amax(cities, axis = 0)
  mx = max(mx[1], mx[2])
  
  cities[:, 1] = (cities[:, 1] - mn) / (mx - mn)
  cities[:, 2] = (cities[:, 2] - mn) / (mx - mn)

  #best_value = -1
  #best_path = []

  for _ in range(iteration_count):
    np.random.shuffle(cities)
    #print(cities)
    
    for c in cities:
      city = c[1:3]
      #print(city)
      closest_node = dist(nn, city).argmin()
      neighbors = find_neighborhood(closest_node, n//10, size * 10)
      nn += neighbors[:, np.newaxis] * learning_rate * (city - nn)
      learning_rate *= 0.99997
      n *= 0.9997
    #cur_path = find_path(nn, cities)
    #cur_val = eval_path(cur_path)
    #print(cur_val)
    #if(cur_val < best_value or best_value == -1):
     # best_value = cur_val
      #best_path = cur_path
    if(n < 1 or learning_rate < 0.001):
      break
    if(_ % 100 == 0):
      cur_path = find_path(nn, cities, real_cities)
      cur_val = eval_path(cur_path)
      print(cur_val, learning_rate, n)
    
    cur_path = find_path(nn, cities, real_cities)
    cur_val = eval_path(cur_path)
    print(cur_val, learning_rate, n)
```
