```python
import random
```

```python
def populate(queens_count, POPULATION_SIZE):
  population = []
  queens = [i for i in range(queens_count)]
  for i in range(0, 2 * POPULATION_SIZE):
    population.append(random.sample(queens, queens_count))
  
  return population
```

fitness returns the number of dead queens

```python
def fitness(gnome):
  t1 = 0
  t2 = 0
  size = len(gnome)
  f1 = [0] * size
  f2 = [0] * size
  for i in range(size):
    f1[i] = gnome[i] - i
    f2[i] = size - gnome[i] - i

  f1.sort()
  f2.sort()

  for i in range(1, size):
    if(f1[i] == f1[i - 1]):
      t1 += 1
    if(f2[i] == f2[i - 1]):
      t2 += 1
  
  return t1 + t2
```

```python
def calculate_fitness(population):
  fit = []
  for i in range(len(population)):
    fit.append(fitness(population[i]))
  return fit
```

```python
def sort_fitness(population):
  fit = calculate_fitness(population);
  sorted_pair = sorted(zip(fit, population))
  ans = []
  for i in range(len(sorted_pair)):
    ans.append(sorted_pair[i][1])
  return ans
```

```python
def selection(population, ELITE_SIZE):
  population = sort_fitness(population)
  return population[:ELITE_SIZE]
```

```python
def single_crossover(parent1, parent2):
  GNOME_SIZE = len(parent1)
  l = random.randint(0, GNOME_SIZE)
  r = random.randint(0, GNOME_SIZE)
  l, r = min(l, r), max(l, r)
  
  child = [0] * GNOME_SIZE

  for i in range(l, r):
    child[i] = parent1[i]
  
  j = 0
  for i in range(0, GNOME_SIZE):
    if(j == l):
      j = r
    if(not(parent2[i] in child)):
      child[j] = parent2[i]
      j += 1

  #if(random.random() <= mutation_rate):
  #  child = random.sample(parent1, len(parent1))

  return child
  
```

```python
def crossover(population, POPULATION_SIZE):
  children = []
  for i in range(POPULATION_SIZE):
    p1 = random.randint(0, len(population) - 1)
    p2 = random.randint(0, len(population) - 1)
    children.append(single_crossover(population[p1], population[p2]))

  return children
```

```python
def local_search(gnome):
  GNOME_SIZE = len(gnome)
  f = fitness(gnome)
  for i in range(GNOME_SIZE - 1):
    tmp = gnome.copy()
    tmp[i], tmp[i + 1] = tmp[i + 1], tmp[i]
    if(fitness(tmp) > f):
      return tmp

  return gnome
```

```python
def mutate(gnome, mutation_rate):
  GNOME_SIZE = len(gnome)
  for i in range(GNOME_SIZE):
    random_pos = random.randint(0, GNOME_SIZE - 1)
    random_pick = random.random()
    if(random_pick <= mutation_rate):
      gnome[i], gnome[random_pos] = gnome[random_pos], gnome[i]
      break
  return gnome
```

```python
def mutate_population(population, mutation_rate):
  for i in range(len(population)):
    population[i] = mutate(population[i], mutation_rate)

  return population
```

```python
def improve(population):
  lst = [i for i in range(len(population))]
  lst = random.sample(lst, 50)
  for i in lst:
    population[i] = local_search(population[i])
  return population
```
